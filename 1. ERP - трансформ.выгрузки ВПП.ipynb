```python
import pandas as pd
```

```python
path_start = 'C:\\Users\\HaritonovDV\\Work_space\\2.Input_data\\'
```

```python
name_file = 'ERP 01.01.19-31.10.19.xlsx'
```

```python
# Открыл файл
df = pd.read_excel(path_start + name_file)
```

```python
# Удалил верхние строки с информацией о фильтрах и периоде выгрузки отчета
df.drop(df.index[0:10], inplace=True)
```

```python
# Смотрим какие колонки "Unnamed: Х" содержат NaN.


# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! 
# При проектировании в выгрузке отчета пустыми были 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 5'
# При запуске скрипта в эксплуатацию на большой выгрузке колонка с ЗК "съехала" и удалению подлежали 
# 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 5'
# поэтому в данном месте нужно глазками проверять и при необходимости менять номер столбца для удаления
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! 


df.head(3)
```

```python
# Удалил совершенно пустые столбцы которые выгружаются в отчете из базы
df.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 5'], axis=1, inplace=True)
```

```python
# Глазками убеждаемся что нужная колонка с номерами ЗК осталась
df.head(3)
```

```python
# Проставил названия колонок
df.columns = ['Реализация/Группа',
              'ЗК',           
              'Производитель/Партномер',
              'Кол-во',
              'Выручка',
              'Себестоимость',
              'Стоимость закупки',
              'Доп.расходы',
              'Валовая прибыль',
              'Рентабельность,%',]
```

```python
df.head(10)
```

```python
# Скопировал столбец ЗК т.к. в нем в строках на первом уровне информации (инф. о реализациях и только о них) стоят значения NaN
# Далее значения NaN заменил на МАРКЕР1 для дальнейшей работы по переносу инф с уровня 1 на уровень 2
df[['Маркер0']] = df[['ЗК']].fillna(value='МАРКЕР1')
```

```python
# Сделал столбец "Маркер" в котором опираясь на столбец "Маркер0" разметил уровни данных
df.loc[df['Маркер0'] == 'МАРКЕР1', 'Маркер'] = '1'
df.loc[df['Маркер0'] != 'МАРКЕР1', 'Маркер'] = '2'
```

```python
# Столбец "Маркер0" свое дело сделал. Удалил
df.drop(['Маркер0'], axis=1, inplace=True)
```

```python
# Создал столбец "Реализация0". В данном столбце скопировал значения с 1 уровня информации. На втором уровне везде проставил "-"
df.loc[df['Маркер'] == '1', 'Реализация0'] = df['Реализация/Группа']
df.loc[df['Маркер'] != '1', 'Реализация0'] = '-'
```

```python
# Перезагрузил индексы чтобы начинаилсь с 1 т.к. дальше в цикле возникала ошибка если запускал со старыми индексами
df = df.reset_index(drop=True)
```

```python
# Создал столбец "Реализация1" в котором буду переносить информацию с уровня 1 на уровень 2
df['Реализация1'] = df['Реализация0']
```

```python
df.head(10)
```

```python
# Цикл который проверяет на каком уровне данных находиться итератор
# Если на первом то переносит в столбец "Реализция1" копию значения с уровня 1
# Если на втором уровне то берет значение из ячейки выше. Получается инф из перевого уровня транслируется на второй
for i in range(len(df.index)-(len(df.index)-1),len(df.index)):
    df['Реализация1'][0] = df['Реализация0'][0]
    if df['Реализация0'][i] == '-':
        df['Реализация1'][i] = df['Реализация1'][i-1]
    else:
        df['Реализация1'][i] = df['Реализация0'][i]
```

```python
df.tail(10)
```

```python
# Обрезал датафрейм удалив строки с первым уровнем информации
df = df.loc[df['Маркер'] != '1']
df
```

```python
# Создал столбец "Производитель" в который извлек инф о производителях
df['Производитель'] = df['Производитель/Партномер'].str.split(', ').str.get(0)
```

```python
# Перезагрузил индексы
df = df.reset_index(drop=True)
```

```python
# Сделал заготовку в виде столбца "Партномер" чтобы далее в него извлечь инф о партномерах
# Сделал технический столбец "Марк_произв". В него далее занесу инф о длинне символов в поле производитель
# чтобы далее циклом и слайсингом отрезать информацию о производителе из столбца "Производитель/Партномер"
# и оставить только инф о партномере
df['Партномер'] = df['Производитель/Партномер']
df['Маркер_произв'] = df['Производитель/Партномер']
```

```python
# Цикл в котором извлекаю инф о партномере
for i in range(0,len(df['Производитель'])):
    df['Маркер_произв'][i] = len(df['Производитель'][i]) + 2
    df['Партномер'][i] = df['Производитель/Партномер'][i][df['Маркер_произв'][i]:len(df['Производитель/Партномер'][i])-1]
```

```python
# Извлек код группы из отчета
# Сделал подготовку для дальнейшего извлечения инф о наименовании клиента по аналогии с предыдущим 
# шагом по извлечению партномеров
df['Код_группы'] = df['Реализация/Группа'].str.split(', ').str.get(0)
df['Клиент'] = df['Реализация/Группа']
df['Маркер_групп'] = df['Реализация/Группа']
```

```python
# Цикл для извлечения инф о наименовании клиента
for i in range(0,len(df['Реализация/Группа'])):
    df['Маркер_групп'][i] = len(df['Код_группы'][i]) + 2
    df['Клиент'][i] = df['Реализация/Группа'][i][df['Маркер_групп'][i]:len(df['Реализация/Группа'][i])]
```

```python
# Сформировал датафрейм в котором расположил колонку в нужном порядке. Выполнена подготовка для выгрузки выходной формы
df2 = df[['Клиент','Код_группы','ЗК','Реализация1','Производитель','Партномер','Кол-во','Выручка',
       'Себестоимость','Стоимость закупки','Доп.расходы','Валовая прибыль','Рентабельность,%']]
```

```python
# Выгрузка выхоной формы
df2.to_excel("C:\\Users\\HaritonovDV\\Work_space\\3.Output_data\\ERP_td.xlsx")
```
